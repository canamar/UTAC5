```python
#Plot graphs
import pandas as pd
import matplotlib.pyplot as plt
path='C:\\Users\\amare\\Documents\\U Texas Austin Data Science\\Course 52\\CC_Default.csv'
df = pd.read_csv(path,header = 1)
#Delete unwanted column ID
del df['ID']
#Plot Limit Balance histogram
plt.hist(df['LIMIT_BAL'], bins=5)
plt.xlabel("LIMIT BALANCE")
plt.ylabel("Count")
plt.suptitle("Limit Balance Histogram")
plt.show()
```

```python
# EDA based on the Titanc model
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas import Series, DataFrame
import matplotlib as mpl
import seaborn as sns 
%matplotlib inline
path='C:\\Users\\amare\\Documents\\U Texas Austin Data Science\\Course 52\\CC_Default.csv'
df = pd.read_csv(path,header = 1)
#Delete unwanted column ID
del df['ID']
df.groupby('EDUCATION')['EDUCATION'].count()
```
